```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.tree import DecisionTreeClassifier

train_data = pd.read_csv('data/train.csv')
test_data = pd.read_csv('data/test.csv')
```

```python
train_data.info()
```

```python
train_data.describe()
```

```python
train_data.head(10)
```

```python
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
train_data[['Age']] = imp.fit_transform(train_data[['Age']])
```

```python
train_data['Embarked'] = train_data['Embarked'].fillna('S')
```

```python
test_data.info()
```

```python
test_data.head(10)
```

```python
test_data.describe()
```

```python
test_data[['Age']] = imp.fit_transform(test_data[['Age']])
```

```python
test_data['Fare'] = test_data["Fare"].fillna(7.75)
```

```python
train_data['Sex'] = train_data['Sex'].replace(['female', 'male'], [0, 1])
test_data['Sex'] = test_data['Sex'].replace(['female', 'male'], [0, 1])
```

```python
train_data['Embarked'] = train_data['Embarked'].replace(['C', 'Q', 'S'], [0, 1, 2])
test_data['Embarked'] = test_data['Embarked'].replace(['C', 'Q', 'S'], [0, 1, 2])
```

```python
features = ["Pclass", "Sex", "Age", "Fare", "Embarked", "SibSp"]
x = train_data[features]
y = train_data['Survived']
X_test = test_data[features]
```

```python
model = DecisionTreeClassifier()
model.fit(x, y)
predictions = model.predict(X_test)
```

```python
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
```

```python
output.to_csv('predictions_3.csv', index=False)
```
